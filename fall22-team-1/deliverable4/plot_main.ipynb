```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
from folium import plugins
from folium.plugins import HeatMap
from matplotlib.pyplot import figure
```

```python
main_df = pd.read_csv('../data/main_dataset_trimmed_geocoded.csv', dtype={'ZIPCODE':str})
```

```python
def read_tuples(string):
    try:
        coords = string.split(',')
        lat = float(coords[0].strip()[1:])
        long = float(coords[1].strip()[:-1])
        coords = (lat, long)
        return coords
    except:
        return None 
```

```python
import ast

# main_df['location'] = main_df['location'].apply(ast.literal_eval)

df = pd.read_csv('../data/main_dataset_trimmed_geocoded.csv', 
                 dtype={'ZIPCODE':str}, converters={'location':read_tuples})
df.head(20)
# main_df[['latitude', 'longitude']] = [ast.literal_eval(x)[:2] for x in main_df['location']]
# main_df = main_df.drop(columns='location')
```

```python
heat_df = df[['location']]
heat_df = heat_df.dropna(axis=0, subset=['location'])

# List comprehension to make out list of lists
heat_data = [[row['location'][0],row['location'][1], 100] for index, row in heat_df.iterrows()]
map = folium.Map(location=[heat_data[0][0], heat_data[0][1]])
# Plot it on the map
HeatMap(heat_data).add_to(map)
map
```
