```python
def read_tuples(string):
    try:
        coords = string.split(',')
        lat = float(coords[0].strip()[1:])
        long = float(coords[1].strip()[:-1])
        coords = (lat, long)
        return coords
    except:
        return None
```

```python
df = pd.read_csv('../data/preprocessed-income-restricted-noncommercial.csv', 
                 dtype={'Zip Code':str}, converters={'coordinates':read_tuples})
df
```
